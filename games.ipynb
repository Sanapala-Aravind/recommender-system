```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Downloads/steam-video-games/steam-200k.csv")
```

```python
df.head()
```

```python
l=df[df['purchase']=='play']
```

```python
l.columns=['user_id','game_name','play','rating','zero']
```

```python
l.drop(['play','zero'],axis=1,inplace=True)
```

```python
l.head()
```

```python
game_table=l.pivot_table(index='user_id',columns='game_name',values='rating')
game_table.head()
```

```python
l.info()
```

```python
ratings=pd.DataFrame(l.groupby('game_name')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(l.groupby('game_name')['rating'].count())
ratings.max()
```

```python
ratings.sort_values('num of ratings',ascending=False).head()
```

```python
ratings.head()
```

```python
df.head()
```

```python
fallout4_user_ratings=game_table['Fallout 4']
fallout4_user_ratings.head()
```

```python
similar_to_fallout4=game_table.corrwith(fallout4_user_ratings)
```

```python
similar_to_fallout4.head()
```

```python
corr_fallout4=pd.DataFrame(similar_to_fallout4,columns=['correlation'])
corr_fallout4.dropna(inplace=True)
corr_fallout4.head()
```

```python
corr_fallout4.sort_values('correlation',ascending=False)
```

```python
corr_fallout4=corr_fallout4.join(ratings['num of ratings'])
corr_fallout4.head()
```

```python
corr_fallout4[corr_fallout4['num of ratings']>100].sort_values('correlation',ascending=False)
```

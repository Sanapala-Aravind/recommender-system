```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("Downloads/anime-recommendations-database/anime.csv")
```

```python
df2=pd.read_csv("Downloads/anime-recommendations-database/rating.csv")
```

```python
df2=df2[df2['rating']!=-1]
df2.info()
```

```python
df2=df2.iloc[0:6337240,:]
df2.info()
```

```python
df1.drop('rating',axis=1,inplace=True)
```

```python
df=df1.merge(df2,on='anime_id')
```

```python
df.info()
```

```python
df.info()
```

```python
anime_table=df.pivot_table(index='user_id',columns='name',values='rating')
anime_table.head()
```

```python
anime_table.info()
```

```python
rating=pd.DataFrame(df.groupby('name')['rating'].mean())
rating.head()
```

```python
rating['num of rating']=pd.DataFrame(df.groupby('name')['rating'].count())
rating.head()
```

```python
rating.sort_values('num of rating',ascending=False).head()
```

```python
shingeki_rating=anime_table['Shingeki no Kyojin']
shingeki_rating.head()
```

```python
similar_to_shingeki=anime_table.corrwith(shingeki_rating)
similar_to_shingeki.head()
```

```python
corr_shingeki=pd.DataFrame(similar_to_shingeki,columns=['correlation'])
corr_shingeki.dropna(inplace=True)
corr_shingeki.head()
```

```python
corr_shingeki.sort_values('correlation',ascending=False).head()
```

```python
corr_shingeki=corr_shingeki.join(rating['num of rating'])
corr_shingeki.head()
```

```python
corr_shingeki[corr_shingeki['num of rating']>100].sort_values('correlation',ascending=False)
```
